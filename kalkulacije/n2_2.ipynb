```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

plt.rcParams['figure.figsize'] = [20, 10]
```

```python
x, x0, y, y0, r, r0, theta, theta0, phi, phi0, R = symbols('x x_0 y y_0 r r_0 theta theta0 phi phi_0 R')
```

```python
expr = (1-r**2)/(1-2*r*cos(phi)+r**2).subs({phi: theta-phi})
expr
```

```python
inti = expr*(sin(theta)**2+sin(theta))
inti
```

```python
res = integrate(inti, (theta, 0, 2*pi))
res
```
