```python
import numpy as np
import matplotlib.pyplot as plt

import plotly
import plotly.graph_objs as go

plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# plot expr 

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sinh(X) * np.sin(Y) / np.sinh(np.pi)

trace1 = go.Surface(x=x, y=y, z=Z, colorscale='Viridis', contours = {
        "x": {"show": True, "start": 0, "end": np.pi, "size": 0.1},
        "y": {"show": True, "start": 0, "end": np.pi, "size": 0.1},
    })
data = [trace1]
layout = go.Layout(
    title='2D Laplace Equation',
    autosize=False,
    width=1500,
    height=1000,
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```

```python
# Pogledamo kaj se dogajha pri x= pi/2 ker graf izgleda simmetricen
x_index = np.argmin(np.abs(x-np.pi/2))

x_slice = Z[x_index, :]

xn = np.linspace(0,np.pi,100 )

plt.plot(xn, x_slice)
# Spodaj je ugibanje
plt.plot(xn, np.sinh(xn)/np.sinh(np.pi), "r.")
```

```python
# Se y=pi
y_index = np.argmin(np.abs(y-np.pi))

y_slice = Z[:,y_index]

yn = np.linspace(0,np.pi,100 )

plt.plot(yn, y_slice)
# Spodaj je ugibanje
plt.plot(yn, np.sin(yn), 'r.')
```
